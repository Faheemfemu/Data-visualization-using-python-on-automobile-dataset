```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
auto=pd.read_csv('automobile_data.csv')
```

```python
auto.head()
```

```python
sns.distplot(auto['city_mpg'])
plt.show
```

```python
sns.jointplot(auto['engine_size'],auto['highway_mpg'])
plt.show
```

```python
sns.jointplot(auto['engine_size'],auto['highway_mpg'], kind="hex")
plt.show()
```

```python
sns.jointplot(auto['engine_size'],auto['highway_mpg'], kind="kde")
plt.show()
```

```python
sns.pairplot(auto[['normalize_losses','engine_size', 'horsepower']])
plt.show()
```

```python
sns.stripplot(auto['fuel_type'], auto['engine_size'])
plt.show()
```

```python
sns.stripplot(auto['fuel_type'], auto['engine_size'], jitter=True)
plt.show()
```

```python
sns.swarmplot(auto['fuel_type'], auto['engine_size'])
plt.show
```

```python
sns.boxplot(auto['num_of_doors'], auto['engine_size'], hue=auto['fuel_type'])
plt.show()
```

```python
sns.barplot(auto['body_style'], auto['engine_size'], hue=auto['engine_location'])
plt.show()
```

```python
sns.countplot(auto['body_style'])
plt.show()
```

```python
sns.pointplot(auto['fuel_system'], auto['engine_size'], hue=auto['num_of_doors'])
plt.show()
```

```python
sns.factorplot(x="fuel_type",
                 y="engine_size",
                 hue="num_of_doors",
                 col="engine_location",
                 data=auto,
                 kind="swarm",)
```

```python
sns.lmplot(x="city_mpg", y="highway_mpg", data=auto)
```

```python
sns.lmplot(x="city_mpg", y="highway_mpg", hue='fuel_type', data=auto)
```
